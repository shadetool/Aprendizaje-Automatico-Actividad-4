```python
import numpy as np
```

```python
# Calculamos el gradiente de la función, es decir, las derivadas parciales respecto a x y y:
def f(x, y):
    return 3*x**2 + 2*x*y + 15*x + 6*y**2 - y
```

```python
# Gradiente de la función
def grad_f(x, y):
    df_dx = 6*x + 2*y + 15
    df_dy = 2*x + 12*y - 1
    return np.array([df_dx, df_dy])
```

```python
# Parámetros del algoritmo de descenso de gradiente
alpha = 0.01  # Tasa de aprendizaje
num_iterations = 1000  # Número de iteraciones
initial_point = np.array([-2, 3])  # Punto inicial
```

```python
# Algoritmo de descenso de gradiente
point = initial_point
for i in range(num_iterations):
    grad = grad_f(point[0], point[1])
    point = point - alpha * grad
    if i % 100 == 0:  # Imprimir cada 100 iteraciones
        print(f"Iteración {i}: x = {point[0]}, y = {point[1]}, f(x,y) = {f(point[0], point[1])}")
```

```python
# Resultado final
print(f"Punto mínimo encontrado: (x, y) = ({point[0]}, {point[1]})")
print(f"Valor mínimo de f(x,y): {f(point[0], point[1])}")
```
